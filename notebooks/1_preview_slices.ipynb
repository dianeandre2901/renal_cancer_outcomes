```python
import pandas as pd
import os
import openslide
from matplotlib import pyplot as plt
```

```python

df = pd.read_csv("/rds/general/user/dla24/home/thesis/TGCA_dataset/tcga_kirc_clinical_parsed.csv")
df.head()

slide_dir = "/rds/general/user/dla24/home/thesis/TGCA_dataset/all_slides"
slide_files = [f for f in os.listdir(slide_dir) if f.endswith(".svs")]

print(f"Found {len(slide_files)} slides")

# Load one slide
slide_path = os.path.join(slide_dir, slide_files[0])
slide = openslide.OpenSlide(slide_path)
print("Slide dimensions:", slide.dimensions)
```

```python
# Show one slide

thumb = slide.get_thumbnail((512, 512))
plt.imshow(thumb)
plt.axis('off')
plt.title(slide_files[0])
plt.show()


for f in slide_files[:5]:  # Show dimensions for first 5 slides
    slide = openslide.OpenSlide(os.path.join(slide_dir, f))
    width, height = slide.dimensions
    print(f"{f}: {width} x {height} pixels")
```
